Datos Generales

```python
sec = pd.read_csv(io.BytesIO(uploaded['Tasa neta de cobertura en secundaria.csv']), encoding="utf8")
sec
```

Matriz de correlación entre los porcentajes de cada año

```python
sec.corr()
```

```python
plt.figure(figsize=(15,6))
sns.heatmap(sec.corr(), annot=True, fmt='.2f')
```
