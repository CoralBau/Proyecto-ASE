```python
cs = pd.read_csv(io.BytesIO(uploaded['Población vulnerable por carencias sociales.csv']), encoding="utf8")
cs
```

```python
plt.figure(figsize=(15,6))
sns.heatmap(cs.corr(), annot=True, fmt='.2f')
```
