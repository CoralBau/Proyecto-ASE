```python
em = pd.read_csv(io.BytesIO(uploaded['Tasa neta de cobertura en educación media superior.csv']), encoding="utf8")
em
```

```python
plt.figure(figsize=(15,6))
sns.heatmap(em.corr(), annot=True, fmt='.2f')
```
