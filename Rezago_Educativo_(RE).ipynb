Datos Generales

```python
re = pd.read_csv(io.BytesIO(uploaded['Rezago Educativo.csv']), encoding="utf8")
re
```

Matriz de correlación entre los porcentajes de cada año 

```python
re.corr()
```

```python
plt.figure(figsize=(15,6))
sns.heatmap(re.corr(), annot=True, fmt='.2f')
```

Gráficas de líneas por año y una comparación de todos los años juntos
