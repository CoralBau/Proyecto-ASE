Gráfica de barras comparativas

```python
# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(20, 6))
 
# set height of bar
d1 = df["Veracruz po"]
d2 = df["Veracruz re"]
d3 = df["Veracruz cs"]
d4 = df["Veracruz es"]
 
# Set position of bar on X axis
br1 = np.arange(len(d1))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
br4 = [x + barWidth for x in br3]
 
# Make the plot
plt.bar(br1, d1, color ='#cda0aa', width = barWidth,
        edgecolor ='grey', label ='Pobreza')
plt.bar(br2, d2, color ="#85c6a9", width = barWidth,
        edgecolor ='grey', label ='Rezago educativo')
plt.bar(br3, d3, color ='#eed4d0', width = barWidth,
        edgecolor ='grey', label ='Carencia social')
plt.bar(br4, d4, color = "#3f3e6fd1", width = barWidth,
        edgecolor ='grey', label ='Educación superior')
 
# Adding Xticks
plt.title('Veracruz',fontsize= 14)
plt.xlabel('Año', fontweight ='bold', fontsize = 15)
plt.ylabel('Porcentaje', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(d1))],
        df['Año'])
 
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15, 5))

plt.plot(df['Año'], df['Veracruz po'], label='Veracruz po')
plt.plot(df['Año'], df['Veracruz re'], label='Veracruz re')
plt.plot(df['Año'], df['Veracruz cs'], label='Veracruz cs')
plt.plot(df['Año'], df['Veracruz es'], label='Veracruz es')

plt.xticks(rotation=45, horizontalalignment='right',fontweight='light', fontsize=20)
plt.xlabel('Año', fontsize=30)
plt.ylabel('Porcentaje', fontsize=30) 
plt.title('Veracruz', fontsize=40) 


plt.grid(True)
plt.legend()

plt.show()
```

Gráfica de barras horizontal

```python
name = df['Veracruz re'].head(12)
price = df['Año'].head(12)

# Figure Size
fig, ax = plt.subplots(figsize =(15, 5))
 
# Horizontal Bar Plot
ax.barh(price,name)
 
# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
 
# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 2)
ax.yaxis.set_tick_params(pad = 7)
 
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
# Show top values
ax.invert_yaxis()
 
# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
    

 
# Add Plot Title
plt.title('Rezago Educativo ',fontsize= 14)
plt.xlabel('Porcentaje')
plt.ylabel('Año')
plt.xticks(rotation=0, horizontalalignment='right',fontweight='light')


# Show Plot
plt.grid(True)


plt.show()
```
