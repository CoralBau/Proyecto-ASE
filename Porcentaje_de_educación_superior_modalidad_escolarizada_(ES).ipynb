```python
es = pd.read_csv(io.BytesIO(uploaded['Porcentaje de educación superior modalidad escolarizada (18-22 años).csv']), encoding="utf8")
es
```

Matriz de correlación entre los porcentajes de cada año

```python
plt.figure(figsize=(15,6))
sns.heatmap(es.corr(), annot=True, fmt='.2f')
```
