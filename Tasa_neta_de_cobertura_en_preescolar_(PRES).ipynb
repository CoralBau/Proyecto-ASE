Datos generales

```python
pres = pd.read_csv(io.BytesIO(uploaded['Tasa neta de cobertura en preescolar.csv']), encoding="utf8")
pres
```

Matriz de correlación entre los porcentajes de cada año

```python
pres.corr()
```

```python
plt.figure(figsize=(15,6))
sns.heatmap(pres.corr(), annot=True, fmt='.2f')
```
